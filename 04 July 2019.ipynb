```python
# Linear Graph

import matplotlib.pyplot as plt

plt.title('Basic linner Graph')
plt.plot([1,2,3],[1,2,3])
plt.show()
```

```python
# Graph to plot distance and time


distance = list(range(2,11,2))
time = list(range(1,6))
#distance=2*time
plt.title('Distance vs Time')
plt.plot(time,distance)
plt.show()
```

```python
# Sample Grid Graph 

distance = list(range(2,11,2))
time = list(range(1,6))
#distance=2*time
plt.title('Distance vs Time')
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(time,distance,'bo')
plt.grid()
plt.show()
```

```python
# Non Linear Graph
t= list( range(1,11))
a= list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.plot(t,v,'go')
plt.grid()
plt.show()
```

```python
## Non Linear Graph anther type graph
t= list( range(1,11))
a= list(range(1,len(t)+1))
v=[a[i]*i for i in range(len(t))]
plt.xlabel('Time')
plt.ylabel('Distance')
plt.plot(t,v)
plt.grid()
plt.show()
```

```python
# Numbers for 
n = list(range(10))
sqn = [i**2 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
# Slop for different cubes

n = list(range(10))
sqn = [i**10 for i in n]
plt.plot(n,sqn)
plt.show()
```

```python
# Voltage -Time
import numpy as np
import matplotlib.pyplot as plt
t= [i/10 for i in range(1,21)]
v= [1+np.sin(2*np.pi*i)for i in t]
plt.title('Time Vs Voltage Graph')
plt.plot(t,v)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.savefig('Images/voltage.png')
plt.show()
```
