```python
# Function to read CSV data into a dataFrame
#return the DataFramw object
filepath = 'DataFiles/Income.csv'
import pandas as pd
def readCSVdata(filepath):
    return pd.read_csv(filepath)

readCSVdata(filepath)
    
    
```

```python
incomedf = readCSVdata(filepath)

def printDataFrameColumns(df):
    columns = df.columns
    for column in columns:
        print(column,end='')
    return
printDataFrameColumns(incomedf)
```

```python
# function to access a row based on a unique

def accessDataFrameRow(df,key):
    #row = df.values[]
    for row in df.values:
        if key in row:
            for item in row:
                print(item, end=' ')
    return

accessDataFrameRow(incomedf,'04000US06')

        
```

```python
# Accessing a Unique Value based on row ,column 
#Income of a state in a given year
def getRowIndex(df,rowkey):
    for i in range(len(df.values)):
        if df.values[i][0]==rowkey or df.values[i][1] == rowkey:
            rowindex = i 
    return rowindex

def getColumnIndex(df,columnkey):
    for i in range(len(df.columns)):
        if df.columns[i] == columnkey:
            columnindex = i
            
    return columnindex   

def valueFromRowColumn(df,rowkey,columnkey):
    rowindex = getRowIndex(df,rowkey)
    columnindex = getColumnIndex(df,columnkey)
    return df.values[rowindex][columnindex]      
                
valueFromRowColumn(incomedf,'04000US06','2009')
```

```python
# Function to update data base on rowkey and

    
def updateDataFomRowColumn(df,rowkey,columnkey,newdata):
    rowindex = getRowIndex(df,rowkey)
    columnindex = getColumnIndex(df,columnkey)
    row = df.values[rowindex]
    row[columnindex] = newdata
    df.loc[rowindex] = row
    return   

updateDataFomRowColumn(incomedf,'Arizona','2007',62993)   
incomedf
```

```python
# Function to add a new row of data to DataFrame

def addRowDataDataframe(df,rowdata):
    lastrowindex = len(df.values)-1
    df.loc[lastrowindex+1] = rowdata
    return

rowdata=['04000US07','Newyark',51755,55319,55734,57014,56134,54283,53367,57020,57528]
addRowDataDataframe(incomedf,rowdata)
incomedf
```

```python
def delectRowDataDataframe(df,rowkey):
    rowindex = getRowIndex(df,rowkey)
    df.drop(rowindex,axis=0)
    return

delectRowDataDataframe(incomedf,1)
incomedf
```
